```python
import numpy as np
import pandas as pd
import os, glob
import pickle
```

```python
# Load all result files

path = '../result/'
all_files = glob.glob(os.path.join(path, "*.csv"))

df_from_each_file = (pd.read_csv(f, names=["time", "nBits", "value", "in_count", "out_count", "count", 
                                           "value_min", "value_max", "in_min", 
                                           "in_max", "out_min", "out_max"]) for f in all_files)
concatenated_df   = pd.concat(df_from_each_file, ignore_index=True)

# Retrieve epoch time by minutes
concatenated_df["time"] = concatenated_df["time"] * 3600
concatenated_df.head()
```

```python
# Load BTCUSD data

btc_usd = pd.read_csv("../data/Bittrex_BTCUSD_1h.csv", header=0)
btc_usd.head()
```

```python
concatenated_df["time"].isin(btc_usd["Unix Timestamp"]).value_counts()
```

```python
result = pd.merge(concatenated_df, btc_usd[["Unix Timestamp", "Open", "Close", "Volume BTC", "Volume USD"]], 
                 left_on = "time", right_on = "Unix Timestamp", how = "inner").drop('Unix Timestamp', 1)

result.head()
```
